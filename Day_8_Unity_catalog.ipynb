```python
from pyspark.sql import SparkSession
from pyspark.sql import Row

# Create a sample DataFrame with columns for products
data = [Row(product_id=1, name="Sample Product", price=9.99)]
df = spark.createDataFrame(data)

# Write to Delta table with mergeSchema=True to create columns
(df.write.format("delta")
    .mode("append")
    .option("mergeSchema", "true")
    .saveAsTable("gold.products"))
```

```python
%sql
CREATE OR REPLACE VIEW gold.top_products AS
SELECT
    product_id,
    name,
    price
FROM gold.products
LIMIT 100;
```
