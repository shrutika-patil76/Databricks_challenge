```python
%python

df = spark.read.csv(
    "/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv",
    header=True,
    inferSchema=True
)

display(df)
```

```python
df.printSchema()
```

```python
df.select(
    "event_time",
    "category_code",
    "brand",
    "price"
).show(10)
```

```python
grouped_df = df.groupBy("category_code") \
    .avg("price") \
    .withColumnRenamed("avg(price)", "avg_price")

grouped_df.show()
```

```python
ordered_df = grouped_df.orderBy("avg_price", ascending=False)

ordered_df.show()
```

```python
from pyspark.sql.functions import col

# Filter records where price > 100
filtered_df = df.filter(col("price") > 100)

# Write filtered data to CSV
filtered_df.write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("/Volumes/workspace/ecommerce/ecommerce_data/output_filtered_csv")
```

```python
ordered_df.write \
    .mode("overwrite") \
    .parquet("/Volumes/workspace/ecommerce/ecommerce_data/output_grouped_parquet")
```
