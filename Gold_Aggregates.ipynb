```python
from pyspark.sql.functions import count, sum, when, col

silver = spark.read.format("delta") \
    .load("/Volumes/workspace/ecommerce/ecommerce_data/delta/silver/events")

gold = silver.groupBy("event_date") \
    .agg(
        count("*").alias("total_events"),
        sum(when(col("event_type") == "purchase", 1).otherwise(0)).alias("purchases")
    )

gold.write.format("delta") \
    .mode("overwrite") \
    .save("/Volumes/workspace/ecommerce/ecommerce_data/delta/gold/daily_metrics")
```
