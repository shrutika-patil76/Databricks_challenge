```python
# ===============================
# DAY 1 CHALLENGE 1: FULL PYSPARK PRACTICE (Error-free)
# ===============================

from pyspark.sql import SparkSession
from pyspark.sql.functions import upper
from pyspark.sql.functions import col, when, avg, max, min, count

# Initialize Spark session (usually already available in Databricks)
spark = SparkSession.builder.appName("Day1_Challenge").getOrCreate()

# 1️⃣ CREATE DATAFRAME FROM INLINE DATA
students_data = [
    ("Shrutika", 25, "Computer"),
    ("John", 30, "Math"),
    ("Alice", 28, "Physics"),
    ("Bob", 22, "Chemistry"),
    ("Emma", 27, "Math"),
    ("David", 31, "Physics"),
    ("Sophia", 24, "Computer")
]

columns = ["Name", "Age", "Department"]

df = spark.createDataFrame(students_data, columns)

print("=== Original DataFrame ===")
df.show()
df.printSchema()

# 2️⃣ DATA EXPLORATION
print("=== Columns & Count ===")
print("Columns:", df.columns)
print("Total rows:", df.count())
```
